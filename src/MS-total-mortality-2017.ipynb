# 2017 Mortality Figure

```python
%%R 

# Custom theme
theme_custom <- theme(plot.title = element_text(size = 24, colour = "grey40"))
```

```python
%%R -w 8 -h 5 -u in -r 200

# Base
q <- ggplot()

# Edit axes scales
q <- (q + scale_x_continuous(limits = c(0, 5)) +
          scale_y_continuous(limits = c(0, 5)))

# Adds fatalities figures
q <- (q + annotate("text", x = 0.3, y = 3, label = "70,237", 
                   size = 25, colour = "#C62F2F", hjust = 0, vjust = 0) +
      
          annotate("text", x = 0.3, y = 2.3, label = "People died from drug overdoses", 
                   size = 10, colour = "grey40", hjust = 0, vjust = 0) +
                  
          annotate("text", x = 0.3, y = 1.80, label = "More than the Vietnam, Afghaistan, and", 
                   size = 8, colour = "grey50", hjust = 0, vjust = 0) +
      
          annotate("text", x = 0.3, y = 1.30, label = "Iraq Wars combined", 
                   size = 8, colour = "grey50", hjust = 0, vjust = 0))
      
# Adds title
q <- q + labs(title = "In the Year 2017 Alone")

# Edit themes
q <- q + theme_void() + theme_custom

q
```
